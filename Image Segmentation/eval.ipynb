```python
arr = np.array([144.15102414,157.14571714,184.01073585])
```

```python
def jaccard_score(arr1,arr2):
    # arr1, arr2: 2-d numpy binary image in 1-d vectors of same length
    a = arr1.astype(int)
    b = arr2.astype(int)
    return sum(a & b) / (1. * sum(a | b))

def dice_score(arr1,arr2):
    a = arr1.astype(int)
    b = arr2.astype(int)
    return sum(a & b) / (0.5 * sum(a | b) + 0.5 * sum(a & b) )
```

# 2. Test on test images

```python

avg = []
avg_crf = []

acc = []
acc_crf = []

recall = []
recall_crf = []

spc = []
spc_crf = []

dice = []
dice_crf = []
# test_list: for test images
# val_list: for val images
for img_, img in enumerate(val_list):
    #input image
    im_jpg = Image.open(data_dir + 'img/' + img)
#     im_jpg = Image.open(data_dir + 'test/' + img)
    im_jpg = im_jpg.resize((500,500), Image.NEAREST)

    in_ = np.array(im_jpg, dtype=np.float32)
    #convert RGB to BGR
    in_ = in_[:,:,::-1]
    #convert [BGR][BGR] to [BB][GG][RR]
    in_ = in_.transpose((2,0,1))
    in_[0] = in_[0] - arr[0]
    in_[1] = in_[1] - arr[1]
    in_[2] = in_[2] - arr[2]
    
    net.blobs['data'].reshape(1, *in_.shape)
    net.blobs['data'].data[...] = in_
    net.blobs['data_dim'].data[...] = np.array([500, 500]).reshape(1, 1, 1, 2)
    
    net.forward()
    #temp
    #net.blobs['upscore'].data[0][0] = net.blobs['upscore'].data[0][0]-2
    
    # Prediction without CRF
    pix_result = net.blobs['upscore'].data[0].argmax(axis=0)

    #label image
    im_png = Image.open('%s/label_zo/%s_Segmentation.png'%(data_f,img[:-4]))
    im_png = im_png.resize((500,500), Image.NEAREST)
    pix_png = np.array(im_png)
    
    
    # Prediction with CRF
    pix_result_crf = net.blobs['upscore_crf'].data[0].argmax(axis=0)
    
    y_true = pix_png.flatten()
    y_pred = pix_result.flatten()
    y_pred_crf = pix_result_crf.flatten()
    
    avg.append(jaccard_score(pix_png.flatten(), pix_result.flatten()))
    avg_crf.append(jaccard_score(pix_png.flatten(), pix_result_crf.flatten()))
    
    acc.append(accuracy_score(pix_png.flatten(), pix_result.flatten()))
    acc_crf.append(accuracy_score(pix_png.flatten(), pix_result_crf.flatten()))

    recall.append(recall_score(pix_png.flatten(), pix_result.flatten()))
    recall_crf.append(recall_score(pix_png.flatten(), pix_result_crf.flatten()))

    spc.append(recall_score(np.ones(y_true.shape) - y_true,np.ones(y_pred.shape) - y_pred))
    spc_crf.append(recall_score(np.ones(y_true.shape) - y_true,np.ones(y_pred_crf.shape) - y_pred_crf))

    dice.append(dice_score(y_true, y_pred))
    dice_crf.append(dice_score(y_true, y_pred_crf))

print("Accuracy \t: %f"%(np.average(acc)))
```
